#### Exploratory Data Analysis
##### Roles to Explore:

```python
import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset
import seaborn as sns
import numpy as np
import ast

# Loading Data set
dataset= load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

##### Filter for India Data Analyst Roles

```python
df_DA_India = df[(df['job_country']=='India') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_plot = df_DA_India['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='job_location',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Analyst in India')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches((12,5))

dict_column = {
'job_work_from_home': 'Work from Home Offered',
'job_no_degree_mention': 'Job Degree Req.',
'job_health_insurance': 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_India[column].value_counts(),labels=df_DA_India[column].value_counts().index.astype(str),startangle=90, autopct='%1.1f%%')
    ax[i].set_title(title)
plt.show()
```

```python
df_plot = df_DA_India['company_name'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='company_name',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('Counts of companies for Data Analyst in India')
plt.xlabel('Number of Jobs')
plt.ylabel('')
```

```python
df_plot = df_DA_India['job_via'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='job_via',hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.title('Counts of Job search Platform for Data Analyst in India')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
