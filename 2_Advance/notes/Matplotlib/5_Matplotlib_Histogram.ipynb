#### Histogram
<div style='font-size:14px'>Histogram are great at showing the distribution of values 

```python
import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset
import numpy as np
import ast

# Loading Data set
dataset= load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_India = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='India')].copy()
```

```python
df_DA_India['salary_year_avg'].plot(kind='hist',bins=30,edgecolor='black')
plt.xlim(0,200000)

ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.title('Distribution of United States Data Analyst Yearly Salaries')
plt.xlabel('Yearly Salary')
plt.ylabel('Number of Jobs')
```
