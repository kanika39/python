so we have a mini little project we're going to work on in this data cleanup. so we have a coworker that wnts access to this data but they have specific requirements first we have arnd 3000 data salary points however the data sets arnd 7000000 so this coworker wants those nan values filled in with something like  median value additionally after we fill that in they want us to go through and remove any duplicate job entries(duplicate job mtlb ek he company mai ek he position ki job baar baar alag alag job_via se entered hai) 

```python
import matplotlib.pyplot as plt
import pandas as pd
from datasets import load_dataset

# Loading Data set
dataset= load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
median_salary_year=df['salary_year_avg'].median()
```

```python
median_salary_hour=df['salary_hour_avg'].median()
```

<div style='font-size:14px'> now we need to fill these median value in place of nan value so we actually going to fill in these nan values withouth replacing the current values in there so for this there is a method in pandas .fillna but whenever we will modify our dataframe we want to keep the original one intact so we will create a new one

```python
df_filled = df

df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)
df_filled.loc[:10,'salary_year_avg':'salary_hour_avg']
```

<div style='font-size:14px'>now 1st part of filling nan is completed now we move to next part to drop duplicate for this pandas have a new method i.e .drop_duplicates

```python
df_unique = df_filled
df_unique=df_unique.drop_duplicates()

print('length of original df: ',len(df_filled))
print('length of drop duplicate df: ',len(df_unique))
print('Rows Dropped: ',len(df_filled)-len(df_unique))
```

```python
df_unique=df_unique.drop_duplicates(subset=['job_title','company_name'])

print('length of original df: ',len(df_filled))
print('length of drop duplicate df: ',len(df_unique))
print('Rows Dropped: ',len(df_filled)-len(df_unique))
```
